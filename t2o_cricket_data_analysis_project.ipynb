```python
import pandas as pd
import json
```

## Match Summary

```python
with open('/content/t20_wc_match_results.json') as f:
  data=json.load(f)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(10)
```

```python
df_match.shape
```

```python
 df_match.rename({'scorecard': 'match_id'}, axis=1, inplace=True)
 df_match.head()
```

```python
match_ids_dict = {}
for index, row in df_match.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']
  match_ids_dict[key1] = row['match_id']
  match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
df_match.to_csv('/content/t20_wc_match_results.json.csv', index = False)
```

## Batting Summary

```python
#batting summary

with open('/content/t20_wc_batting_summary.json') as f:
  data=json.load(f)

  all_records = []

  for rec in data:
    all_records.extend(rec['battingSummary'])
      #extend means adding one array after the other
df_batting = pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(
    lambda x: "out" if len(x) > 0 else "not_out"
    )
##changes are done to make dismissal column as out or not out

df_batting.head(10)
```

```python
df_batting.drop(columns=["dismissal"], inplace = True)
df_batting.head(10)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head(5)


```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('/content/t20_wc_batting_summary.json.csv', index = False)
```

## Bowling Summary

```python
with open('/content/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head(3)
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dict)
df_bowling.head(5)
```

```python
df_bowling.to_csv('/content/t20_wc_bowling_summary.json.csv', index = False)
```
